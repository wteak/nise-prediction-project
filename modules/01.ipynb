```python
import pandas as pd
import time
import os

# [Module 1] 데이터 로더 및 포맷 변환기
# 목적: 대용량 CSV(1900만 행)의 로딩 속도 개선 (6분 -> 7초)

CSV_TRAIN = 'train.csv'
CSV_TEST = 'test.csv'
FEATHER_TRAIN = 'train_original.feather'
FEATHER_TEST = 'test_original.feather'

print("--- [Step 1] CSV to Feather 변환 시작 ---")
start = time.time()

# 1. Train 데이터 변환
if not os.path.exists(FEATHER_TRAIN):
    print(f"'{CSV_TRAIN}' 로드 중...")
    df_train = pd.read_csv(CSV_TRAIN, parse_dates=['time'])
    df_train.to_feather(FEATHER_TRAIN)
    print(f" -> Train 변환 완료. ({time.time() - start:.2f}초)")
else:
    print(f" -> '{FEATHER_TRAIN}' 파일이 이미 존재합니다.")

# 2. Test 데이터 변환
if not os.path.exists(FEATHER_TEST):
    print(f"'{CSV_TEST}' 로드 중...")
    df_test = pd.read_csv(CSV_TEST, parse_dates=['time'])
    df_test.to_feather(FEATHER_TEST)
    print(f" -> Test 변환 완료.")
else:
    print(f" -> '{FEATHER_TEST}' 파일이 이미 존재합니다.")

print("데이터 준비 완료.")
```
